```python
data = {'a' : [1,2,3,4], 'b' : [0, 1, 0.1, 0.2], 'sub':['1st', '2nd', '3rd', '4th']}
data
```

```python
df3 = pd.DataFrame(data)
df3
```

```python
df3 = pd.DataFrame(data, columns = ['sub','a','b']) # 데이터의 열의 순서를 변경 가능
df3
```

```python
df3['c'] = np.nan # 새로운 열, 값 할당 가능
df3
```

```python
df3.loc[1, ['c']] = 2 # 위치 값도 접근 가능 함
df3
```

```python
df3['d'] = df3['a'] + df3['b'] # 기존 열 값들을 이용해서 간단히 열 추가 가능
df3
```

```python
df3.loc[4] = np.nan #행 추가는 새로운 인덱스(=행 정보)를 선언해서 대입해준다.
df3
```

```python
import pandas as pd

col_names = ['스트레스', '스트레스남학생', '스트레스여학생', '우울감경험률', '우울남학생', '우울여학생', '자살생각율', '자살남학생', '자살여학생']

raw_data = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/머신러닝의 이해/teenage_mental.xls", header=1, usecols = "C:K", names = col_names)
#header = 1은 맨 윗 줄, Column C에서 Column K까지만 필요한 부분 가져온다.
raw_data
```

```python
raw_data.loc[1] = 100. - raw_data.loc[0] # 응답이 아니다의 데이터를 추가. 행에 접근할 때는 .loc 사용
raw_data
```

```python
raw_data['응답'] = ['그렇다', '아니다'] #응답 column을 생성
raw_data.set_index('응답', drop = True, inplace = True) #응답 column이 아니라 인덱스(행)으로 바꾸겠다.

raw_data
```

```python
f, ax = plt.subplots(figsize=(10,8))

raw_data['스트레스'].plot.pie(explode = [0, 0.02])
ax.set_title('스트레스를 받은적 있다') # 제목 추가
ax.set_ylabel('') # y축 제목 삭제
plt.show()
```

```python
f, ax = plt.subplots(1, 3, figsize = (16, 8))

raw_data['스트레스'].plot.pie(explode=[0, 0.02], ax = ax[0], autopct = '%1.1f%%') # 첫 번째 이미지는 ax[0], 이후 이미지는 ax[1], ax[2]로 된다. 앞 쪽에 한칸의 공백을 두고 정렬. 소수점 첫 째 자리까지 그리고 실수형으로 출력한다.
ax[0].set_title('스트레스를 받은적 있다')
ax[0].set_ylabel('')


raw_data['우울감경험률'].plot.pie(explode=[0, 0.02], ax = ax[1], autopct = '%1.1f%%')
ax[1].set_title('우울증을 경험한적 있다')
ax[1].set_ylabel('')


raw_data['자살생각율'].plot.pie(explode = [0, 0.02], ax = ax[2], autopct = '%1.1f%%')
ax[2].set_title('자살을 고민한 적 있다')
ax[2].set_ylabel('')

plt.show()
```
