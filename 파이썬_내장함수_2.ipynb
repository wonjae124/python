```python
# repr은 representation의 약어, printable representation

class Person(object):
  def __init__(self,name,age):
    self.name = name
    self.age = age
  
  def __repr__(self):
    return "<이름:%s, 나이:%s>"%(self.name,self.age)

def keyage(person):
  return person.age
```

```python
# 파이썬에는 for문에 들어 있는 iterable object라고 한다.

class MyCounter(object):
  def __init__(self, low, high):
    self.current = low
    self.high = high

# iteration 객체
  def __iter__(self):
    return self
  
# 다음 반복을 위해 값을 반환하는 객체
  def __next__(self):
    if self.current>self.high:
      raise StopIteration
    else:
      self.current += 1
      return self.current-1

for i in MyCounter(1,10):
    print(i, end=' ')
```

```python
# generator는 yield 사용
# iterator와 마찬가지로 for문에서 사용
def myGenerator():
  yield 'first'
  yield 'second'
  yield 'third'

for word in myGenerator():
  print(word)
```

```python
# Closure, 함수에 의해 반환
def addNumber(fixedNum):
  def add(number):
    return fixedNum + number
  return add
# addnumber(A), fixedNumb에 A 할당으로 A+number 반환
# func(B), number에 B 할당으로 A+B 반환 
func = addNumber(10)
print(func(20)) 
```

```python
# 연산자 오버로딩
# 인스턴스 객체끼리 서로 연산할 수 있게끔 기존 연산자의 기능을 중복으로 재정의
# 사칙연산, 비트 이동, 비트 연산자

class Point:
  def __init__(self, x=0, y=0):
    self.x = x
    self.y = y
  
# 오버로딩! 재정의!
  def __add__(self,other):
    x = self.x + other.x
    y = self.y + other.y
    return Point(x,y)

# 반환시 주소가 아니라, 문자열 및 숫자가 조합된 표현으로 저장해두기
  def __str__(self):
    return 'Point('+str(self.x)+','+str(self.y)+')'


p1 = Point(1,2)
p2 = Point(3,4)
print(p1+p2)
```

```python
# 키워드 모듈
# 리스트 내부의 요소를 파악 가능
import keyword
name = input("변수 이름을 입력하시오:")

if keyword.iskeyword(name):
  print(name,"은 예약어임.")
  print("아래는 키워드의 전체 리스트임: ")
  print(keyword.kwlist)
else:
  print(name, "은 사용할 수 있는 변수이름임.")
```
